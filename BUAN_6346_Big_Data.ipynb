```python
import pandas as pd
import category_encoders as ce


# create a leave-one-out encoder object
looe = ce.LeaveOneOutEncoder()

# apply target encoding to columns 'col1' and 'col2'
geo[['Model_te', 'driverid_te', 'city_te']] = looe.fit_transform(geo[['Model', 'driverid', 'city']], geo['event_ind'])

# display the encoded dataframe
print(geo)
```

```python
target_col = geo['event_ind']
predictor_cols = geo[[ 'velocity','Riskfactor','Model_te', 'city_te']]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(predictor_cols, target_col, test_size=0.2, random_state=1)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()

# scale the data
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
X_train
```

```python
# Training the Logistic Regression model on the Training set

from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state = 0)
result = classifier.fit(X_train, y_train)
```

```python
# Predicting the Test set results

y_pred = classifier.predict(X_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.to_numpy().reshape(len(y_test),1)),1))
```

```python
# Making the confusion Matrix

from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```

```python
# import the necessary libraries
from sklearn.linear_model import LogisticRegression
import statsmodels.api as sm

# define the list of coefficients
coefficients = ['velocity', 'Riskfactor', 'Model_te', 'city_te']

# create and train a logistic regression model
logreg = LogisticRegression()
logreg.fit(X_train, y_train)

# get the coefficients and intercept for the trained model
coef = logreg.coef_
intercept = logreg.intercept_

# fit the logistic regression model using statsmodels to get the p-values
logit_model = sm.Logit(y_train, X_train)
result = logit_model.fit()

# loop through the list of coefficients and print the relevant information
for i in range(len(coefficients)):
    print("p-value for {}: {}".format(coefficients[i], result.pvalues[i]))

print("\n")

# loop through the list of coefficients and print the relevant information
for i in range(len(coefficients)):
    print("Coefficient for {}: {}".format(coefficients[i], coef[0][i]))

print("\n")

# loop through the list of coefficients and print the relevant information
for i in range(len(coefficients)):
    print("Exp of Coefficient for {}: {}".format(coefficients[i], np.exp(coef[0][i])))

print("\n")

# loop through the list of coefficients and print the relevant information
for i in range(len(coefficients)):
    if i >= 2:
        print("1/Exp of Coefficient for {}: {}".format(coefficients[i], 1/np.exp(coef[0][i])))

# print the intercept
print("\nIntercept:", intercept)
```

```python
import statsmodels.api as sm

y = geo['event_ind']
X = geo[['velocity','Riskfactor','Model_te', 'city_te']]

# fit the logistic regression model
logit_model = sm.Logit(y, X).fit()

# print the summary of the model
print(logit_model.summary())

# print the p-values for each coefficient
print(logit_model.pvalues)

# print the R-squared value
print("R-squared: ", logit_model.prsquared)
```
